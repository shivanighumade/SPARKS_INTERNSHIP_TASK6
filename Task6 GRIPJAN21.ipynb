<h2> PREPARING THE DATA

```python
target = data['Species']
```

```python
# Drop Country and Postal code columns
data1 = data.copy()
data1 = data.drop('Species', axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# Divide the data into "attributes" (inputs) and "labels" (outputs).
X = data.iloc[:, [0,1,2,3]].values  

le = LabelEncoder()
data['Species'] = le.fit_transform(data['Species'])
y = data['Species'].values #labels
```

<h2> TRAIN-TEST SPLIT

<h3>SPLITTING THE DATA

```python
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0) 
```

<h2> BUILDING THE MODEL

```python
# Defining the decision tree algorithm
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf.fit(X_train,y_train)
```

```python
print("Training complete.")
```

```python
# Calculating the accuracy of the model
clf.score(X_train, y_train)
```

```python
clf.score(X_test,y_test)
```

<h2> VISUALIZING THE TREE

```python
from sklearn.tree import plot_tree
```

```python
# Visualize the graph
plt.figure(figsize=(20,20))

dec_tree = plot_tree(clf, feature_names = data1.columns,
                    class_names= target.values, filled=True, precision=4, rounded=True);
```
